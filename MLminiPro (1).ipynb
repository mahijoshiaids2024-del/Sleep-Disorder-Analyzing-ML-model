```python
df = df.drop(["Patient_ID", "OCR_Extracted_Text"], axis=1)
```

```python
import sklearn as sk
from sklearn.preprocessing import LabelEncoder,StandardScaler
from sklearn.model_selection import train_test_split
from sklearn.linear_model import LogisticRegression
from sklearn.ensemble import RandomForestClassifier
from sklearn.svm import SVC
from sklearn.metrics import confusion_matrix,accuracy_score,precision_score,recall_score,f1_score,classification_report
```

```python
le = LabelEncoder()
df["Gender"]=le.fit_transform(df["Gender"])
df["Sleep_Disorder_Type"]=le.fit_transform(df["Sleep_Disorder_Type"])
```

```python
df
```

```python
x=df.drop(columns="Diagnosis_Confirmed")
```

```python
y=df["Diagnosis_Confirmed"]
```

```python
x_train,x_test,y_train,y_test=train_test_split(x,y,train_size=0.80,random_state=50)
```

```python
scaler=StandardScaler()
x_train_scale=scaler.fit_transform(x_train)
x_test_scale=scaler.transform(x_test)
```

### Logistic Regression

```python
model=LogisticRegression()
```

```python
model.fit(x_train_scale,y_train)
```

```python
y_pred = model.predict(x_test_scale)
```

```python
acc=accuracy_score(y_test,y_pred)
```

```python
acc
```

```python
cm=confusion_matrix(y_test,y_pred)
```

```python
cm
```

```python
precision=precision_score(y_test,y_pred)
```

```python
precision
```

```python
recall=recall_score(y_test,y_pred)
```

```python
recall
```

```python
f1=f1_score(y_test,y_pred)
```

```python
f1
```

```python
report=classification_report(y_test,y_pred)
```

```python
print(report)
```

```python
import matplotlib.pyplot as plt
```

```python
import seaborn as sns
```

```python
# Distribution of Sleep Disorder Types
plt.figure(figsize=(8,5))
sns.countplot(data=df, x="Sleep_Disorder_Type", palette="Set2")
plt.title("Distribution of Sleep Disorder Types")
plt.xlabel("Type of Sleep Disorder")
plt.ylabel("Number of Patients")
plt.xticks(rotation=20)
plt.tight_layout()
plt.show()
```

```python
#Gender Distribution per Disorder
plt.figure(figsize=(8,5))
sns.countplot(data=df, x="Sleep_Disorder_Type", hue="Gender", palette="pastel")
plt.title("Gender Distribution per Sleep Disorder Type")
plt.xlabel("Sleep Disorder Type")
plt.ylabel("Count")
plt.xticks(rotation=20)
plt.tight_layout()
plt.show()
```

```python
#Correlation Heatmap for Numeric Features
plt.figure(figsize=(6,4))
sns.heatmap(df.select_dtypes("number").corr(), annot=True, cmap="coolwarm", linewidths=0.5)
plt.title("Correlation Heatmap of Numeric Features")
plt.tight_layout()
plt.show()
```

```python
# Relationship between AHI Score and SaO2 Level
plt.figure(figsize=(7,5))
sns.scatterplot(data=df, x="AHI_Score", y="SaO2_Level",
                hue="Sleep_Disorder_Type", style="Gender", alpha=0.8, s=70)
plt.title("AHI Score vs SaO2 Level by Disorder Type")
plt.xlabel("AHI Score (Apnea–Hypopnea Index)")
plt.ylabel("SaO2 Level (Oxygen Saturation)")
plt.legend(title="Sleep Disorder Type")
plt.tight_layout()
plt.show()
```

```python
# Age Distribution per Sleep Disorder Type
plt.figure(figsize=(7,5))
sns.boxplot(data=df, x="Sleep_Disorder_Type", y="Age", palette="Set3")
plt.title("Age Distribution per Sleep Disorder Type")
plt.xlabel("Sleep Disorder Type")
plt.ylabel("Age (years)")
plt.xticks(rotation=20)
plt.tight_layout()
plt.show()
```
